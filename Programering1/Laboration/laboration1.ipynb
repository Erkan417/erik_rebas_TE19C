```python
import math 
import matplotlib.pyplot as plt
import random

nummer = 0

for i in range(100000): #simulera 100 punkter

    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    r = math.sqrt(x**2 + y**2) #räknat ut radien
    if r<1:
        nummer += 1
    
svar= nummer/100000
print(svar)
    

```

```python
import math 
import matplotlib.pyplot as plt
import random

nummer = 0 #nummer startar på 0

for i in range(10000000): #simulera antalet punkter

    x = random.uniform(-1, 1) #skriva up värdena för variabeln x
    y = random.uniform(-1, 1) #skriva up värdena för variabeln y
    r = math.sqrt(x**2 + y**2) #räknat ut radien
    if r<1: 
        nummer += 1 #om radien är mindre än 1 så blir nummer adderat med 1
    
svar= nummer/10000000
print(svar*4) #svaret närmar sig pi alltmer efter hur många punkter man simulerar vet inte riktigt varför
```

```python
import math 
import matplotlib.pyplot as plt
import random


for i in range(1000): #simulera antalet punkter

    x = random.uniform(-1, 1) #skriva up värdena för variabeln x
    y = random.uniform(-1, 1) #skriva up värdena för variabeln y
    r = math.sqrt(x**2 + y**2) #räknat ut radien med hjälp av pythagoras sats
    if r<1: 
        plt.plot(x, y, '*b')
    else:
        plt.plot(x, y, '*r')
    
```

```python
import math 
import matplotlib.pyplot as plt
import random

nummer = 0

for i in range(1000): #simulera antalet punkter

    x = random.uniform(-1, 1) #skriva up värdena för variabeln x
    y = random.uniform(-1, 1) #skriva up värdena för variabeln y
    r = math.sqrt(x**2 + y**2) #räknat ut radien med hjälp av pythagoras sats
    if r<0.1: 
        nummer += 1

svar= nummer/1000
print(svar)

#med radien på 1 fick jag ungefär 78% träffar
#med radien på 0.5 fick jag ungefär 20% träffar
#med radien på 0.1 fick jag ungefär 1% träffar
#ju lägre radien är desto mindre andel träffar
```
