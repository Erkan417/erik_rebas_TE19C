```python
import numpy as np 
import matplotlib.pyplot as plt 
x = np.arange(0,11) # kan använda linspace för att få en mindre hackig graf

def f(x):
    y = 3*x + 1
    return y


plt.plot(f(x))
plt.title("graf")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
